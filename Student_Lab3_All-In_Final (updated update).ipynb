# **Encode the variables in x**

```python
print(x.shape)
```

```python
from sklearn.preprocessing import LabelEncoder 
car_price_le = LabelEncoder()
x[:, 2] = car_price_le.fit_transform(x[:, 2])
```

```python
with np.printoptions(threshold=np.inf):
  print(x)
```

```python
#x = pd.get_dummies(x), print(x)
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
OneEn =OneHotEncoder()
ct = ColumnTransformer(transformers=[('encoder', OneEn, [3,4,5,6,7,8,14,15,17])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
with np.printoptions(threshold=np.inf):
  print(x)
```

**Splitting the dataset into the Training set and Test set**



```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

```python
print(x_train)
```

```python
print(y_train)
```

# **Train the Multiple Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#y_test = y_test.to_numpy(), print(y_test)
print(x_test)
```

```python
print(y_test)
```

# Predicting the Test set results

```python
y_pred = regressor.predict(x_test)
#np.set_printoptions(suppress=True) gives shotens output in exponential form
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```
