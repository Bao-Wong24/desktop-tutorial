# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Import the Dataset

```python
fish_dataset = pd.read_csv('Fish Market.csv')
x = fish_dataset.iloc[:, 1:].values
y = fish_dataset.iloc[:, 0].values
fish_dataset.head()
```

```python
pd.set_option('display.max_rows', None)
print(fish_dataset.describe())
```

```python
with np.printoptions(threshold=np.inf):
  print(x)
```

```python
print(x.shape)
```

```python
with np.printoptions(threshold=np.inf):
  print(y)
```

# Data Preprocessing Stage 

# Encoding Categorical Features (columns) into numeric values for use in regression

```python
print(y)
y = y.reshape(-1, 1)
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
with np.printoptions(threshold=np.inf):
  print(y)
```

# Split the Dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
fish_x_train, fish_x_test, fish_y_train, fish_y_test = train_test_split(x, y, test_size=0.3, random_state = 1)
np.set_printoptions(suppress=True)
print(fish_x_train)
```

# Machine Learning Stage 

# Create the Polynomial model on the training set

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(fish_x_train)
```

# Train Model

```python
lin_reg = LinearRegression()
lin_reg.fit(x_poly, fish_y_train)
```

# Test the Model

```python
print(fish_x_test)
```

# Compare the test set predictions with the actual values

```python
print(fish_y_test)
```

```python
y_poly_pred = lin_reg.predict(poly_reg.fit_transform(fish_x_test))
print(y_poly_pred)
```

```python
np.set_printoptions(precision=2)
print(np.concatenate((y_poly_pred.reshape(len(y_poly_pred), 1), fish_y_test.reshape(len(fish_y_test), 1)), 1))
```
